```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
gnb = GaussianNB()
```

```python
X,Y = make_classification(n_samples=200,n_features=2,n_informative=2,n_redundant=0,random_state=4)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
print(X[0])
print(X.shape)
```

```python
# Train the classifier
gnb.fit(X,Y)
```

```python
gnb.score(X,Y)
```

```python
y_pred = gnb.predict(X)
print(y_pred)
```

```python
print(Y)
```

```python
acc = np.sum(y_pred==Y)/X.shape[0]
print(acc)
```

#### Generate Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(Y,y_pred)
print(cnf_matrix)
```

#### Visualize Confusion Matrix

```python
import confusion_matrix_plot as cplot
```

```python
cplot.plot_confusion_matrix(cnf_matrix,[0,1],normalize=False,title="Confusion Matrix",cmap=plt.cm.Accent)
```
