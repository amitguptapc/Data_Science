```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
plt.style.use('seaborn')
```

LINE PLOT

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.plot(x,y2,color='green',label="Kiwi",linestyle='dashed',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruit over time")
plt.legend()
plt.show()
```

```python
prices=np.array([1,2,3,4,5])**3
plt.plot(prices) # takes indices as x and parameter as y
plt.show()
```

SCATTER PLOT

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y1,color='red',label="Apple")
plt.scatter(x,y2,color='green',label="Kiwi",marker='^')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Prices of fruit over time")
plt.legend()
plt.show()
```

BAR GRAPH

```python
plt.style.use("dark_background")
xcord=np.array([0,1,2])*2
plt.bar(xcord,[10,20,32],width=0.5,label='Current Year',tick_label=['Gold',"Platinum","Silver"],color="Blue")
plt.bar(xcord+0.5,[20,10,12],width=0.5,label='Next Year',color="orange")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,50)
plt.legend()
plt.show()
```

PIE CHART

```python
plt.style.use("seaborn")
subjects=["Maths","Chem","Physics","English"]
weightage=[20,10,15,5]
plt.pie(weightage,labels=subjects,explode=(0,0,0,0.3),shadow=True,autopct='%1.1f%%')
plt.title("Subjects")
plt.show()
```

NORMAL DISTRIBUTION

```python
# Standard Normal Distribution has mean = 0 and standard deviation = 1
```

```python
Xsn = np.random.randn(100)
sigma = 8
u=70
X = np.round(Xsn*sigma + u)
X2= np.round(Xsn* 15+ 40)
print(X)
```

HISTOGRAM

```python
plt.hist(X,alpha=0.8,label="Maths")
plt.hist(X2,alpha=0.8,label="Physics")
plt.xlabel("Marks Range")
plt.ylabel("Frequency")
plt.legend()
plt.title("Histogram")
plt.show()
```
