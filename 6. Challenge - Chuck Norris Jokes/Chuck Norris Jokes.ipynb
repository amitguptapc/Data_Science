```python
import requests
import json
import pandas as pd
```

```python
f=open("Test/ID.csv",'r')
lst=f.readlines()
lst=[int(value[:-1]) for value in lst[1:]]
print(lst)
f.close()
```

```python
with open("output.csv","w") as f:
    f.write("ID,Joke\n")
    result=""
    for idx in lst:
        url="http://api.icndb.com/jokes/{}".format(idx)
        response=requests.get(url)
        data=json.loads(response.content)
        result=data['value']['joke']
        if ',' in result:
            result='{},"'.format(idx)+result+'"\n'
        else:
            result="{},".format(idx)+result+"\n"
        f.write(result)
        print(result)
```

```python
#Alternate Method
all_jokes=[]
for idx in lst:
        url="http://api.icndb.com/jokes/{}".format(idx)
        response=requests.get(url)
        data=json.loads(response.content)
        result=data['value']['joke']
        all_jokes.append(result)
```

```python
df=pd.DataFrame({"ID":lst,"Joke":all_jokes})
print(df)
```

```python
df.to_csv("outputAlt.csv",index=False)
```
