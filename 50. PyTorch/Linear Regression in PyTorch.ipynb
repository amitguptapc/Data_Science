```
import numpy as np
from torch import nn
import torch
from matplotlib import pyplot as plt
```

```
data = np.random.multivariate_normal([3,4],[[1,0.7],[0.7,1]],size=200)
x = data[:,0]
y = data[:,1]
```

```
plt.style.use('seaborn')
plt.scatter(x,y)
plt.show()
```

```
x,y = torch.tensor(x),torch.tensor(y)
m = nn.Parameter(data=torch.randn(1))
c = nn.Parameter(data=torch.randn(1))
```

```
y_pred = m*x + c
plt.scatter(x,y)
plt.plot(x,y_pred.detach(),color='g')
plt.show()
```

```
def mse(y_pred,y_true):
  return ((y_pred-y_true)**2).mean()
```

```
mse(y_pred,y)
```

```
lr = 1e-3
for i in range(100):
  y_pred = m*x+c
  loss = mse(y_pred,y)
  loss.backward()
  print(loss)
  with torch.no_grad():
    m -= lr*m.grad
    c -= lr*c.grad
    m.grad.zero_()
    c.grad.zero_()
```
