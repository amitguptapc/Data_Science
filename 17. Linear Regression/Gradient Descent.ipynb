## Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
# Given a function f(x), find value of x that minimizes f(x)
```

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
x = 0 # Initially
lr = 0.1 # learning rate
error = []
plt.plot(X,Y)
while True:
    grad = 2*(x-5)
    x_new = x - lr * grad
    y = (x_new-5)**2
    error.append(y)
    if (x_new-x)**2 < 0.00001:
        break
    x = x_new
    plt.scatter(x,y)
plt.show()
```

```python
# Plotting the error
plt.plot(error)
plt.show()
```
