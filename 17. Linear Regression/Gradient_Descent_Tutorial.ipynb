## What is Gradient Descent in Machine Learning? 
- Optimization Algorithm
- Based upon First Order Derivatives
- Most widely used technique starting from Linear Regression to Deep Learning
- Other optmization methods - Closed Form Solution, Second Order Derivates etc

### Let Us see a very Simple Case!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
print(x)
```

```python
def lossFn(x):
    return (x-5)**2
```

```python
loss = lossFn(x)
```

```python
plt.style.use("seaborn")
plt.plot(loss)
plt.show()
```

```python
def gradDescent():
    x_ = 0 #or any random value
    learning_rate = 0.1
    print(x_)
    
    plt.plot(loss)
    for i in range(30):
        grad = 2*(x_-5)
    
        x_ = x_ - learning_rate*grad
        l = lossFn(x_)
        print("X_ : %.3f Loss:  %.3f  "%(x_,l))
        plt.scatter(x_,l)
        
        
        
        
        
    
    
```

```python
gradDescent()
```
