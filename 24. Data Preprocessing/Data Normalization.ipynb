```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```python
dfx = pd.read_csv("Datasets/linearX.csv")
dfy = pd.read_csv("Datasets/linearY.csv")
x = dfx.values
y = dfy.values
```

```python
# Normalized data has mean = 0 and standard deviation = 1
x1 = (x-x.mean())/x.std()
plt.scatter(x,y,label="Raw Data")
plt.scatter(x1,y,color="red",label="Normalized Data")
plt.legend()
plt.show()
```
