```python
import numpy as np
import matplotlib.pyplot as plt
```

UNIVARIATE NORMAL DISTRIBUTION

```python
u = 5
sigma = 2
vals = u + sigma * np.random.randn(1000)
print(vals.shape)
print(vals)
```

```python
plt.style.use('seaborn')
plt.hist(vals,50)
plt.show()
```

```python
vals1=np.round(vals)
z=np.unique(vals1,return_counts=True)
print(z)
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

MULTIVARIATE NORMAL DISTRIBUTION

```python
# Bivariate
mean1 = np.array([0.0,0.0])
cov1 = np.array([[1,0],[0,3]])

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist1 = np.random.multivariate_normal(mean1,cov1,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
```

```python
plt.scatter(dist1[:,0],dist1[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
